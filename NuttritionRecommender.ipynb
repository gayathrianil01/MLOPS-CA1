```python
import random
```

```python
class NutritionRecommender:
    def __init__(self, age, weight, dietary_preferences):
        self.age = age
        self.weight = weight
        self.dietary_preferences = dietary_preferences

        self.meal_options = {
            "vegetarian": ["lentil soup", "vegetable curry", "quinoa salad"],
            "vegan": ["vegetable stir-fry", "hummus and pita bread", "fruit smoothie"],
            "non-vegetarian": ["chicken breast with vegetables", "fish with rice", "steak with salad"],
            "gluten-free": ["salmon with sweet potato", "chicken stir-fry with rice noodles", "vegetable soup"],
            "low-carb": ["grilled chicken with asparagus", "salmon with cauliflower mash", "spinach salad with feta cheese"]
        }

    def recommend_meal(self, meal_time):

        if self.age < 30:
            pass
        elif self.age > 60:
            pass
        if self.weight < 120:
            pass
        elif self.weight > 200:
            pass

        meal_options = self.meal_options[self.dietary_preferences]

        meal = random.choice(meal_options)

        return meal

    def recommend_meal_plan(self):

        breakfast = self.recommend_meal("breakfast")
        lunch = self.recommend_meal("lunch")
        dinner = self.recommend_meal("dinner")

        return breakfast, lunch, dinner

user_data = {
    "age": 35,
    "weight": 150,
    "dietary_preferences": "vegetarian"
}

nutrition_recommender = NutritionRecommender(**user_data)
meal_plan = nutrition_recommender.recommend_meal_plan()
print("Meal Plan:")
print("Breakfast:",meal_plan[0])
print("Lunch:",meal_plan[1])
print("Dinner:",meal_plan[2])
```

```python
import pandas as pd
import numpy as np
import random

# Define the number of rows you want in the CSV
num_rows = 100

# Create random data
data = {
    'User id': [f'UID{str(i).zfill(4)}' for i in range(1, num_rows+1)],
    'Age': np.random.randint(18, 65, size=num_rows),  # Age between 18 and 65
    'Weight': np.random.uniform(50.0, 120.0, size=num_rows),  # Weight between 50kg and 120kg
    'Dietary_Preferences': random.choices(['Vegan', 'Vegetarian', 'Non-Vegetarian', 'Keto', 'Paleo'], k=num_rows),
    'Recommended_Calories': np.random.uniform(1500, 3000, size=num_rows)  # Recommended Calories between 1500 and 3000
}

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('random_user_data.csv', index=False)

print("CSV file created successfully!")
```
